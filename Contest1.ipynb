```python
# Data Analysis
import pandas as pd

# Data Visualization
from matplotlib import pyplot as plt
from matplotlib.figure import Figure

# Utilities
import numpy as np
import math
import Constants as const
from sklearn import linear_model
from sklearn.impute import KNNImputer

# print(const.const_dict["MMSE_MAX"])
```

```python
#Load dataset
df = pd.read_excel('TrainingSet.xlsx')
# df = pd.read_excel('TestSet.xlsx')
df
```

```python
# Convert Sex attribute in UpperCase
df['Sesso'] = df['Sesso'].str.upper()

# Remove CAP attribute
df = df.drop(columns =['CAP'])

# Remove unlabeled instances
df = df.loc[df["Patologia"] == df["Patologia"]]

dim_df = len(df['ID'])
```

```python
df_clean = pd.DataFrame(columns = df.columns[0:4], data = df.iloc[:,0:4])
df_clean
```

```python
def lin_reg(feature, scolarita, eta, raw_score):
    # Create linear regression object
    regr = linear_model.LinearRegression(normalize = True)
    # Train the model using the training sets
    df_test = df.loc[(df["Scolarita"] == df["Scolarita"]) & (df["Eta"] == df["Eta"]) & (df[feature+"_PG"] == df["MMSE_PG"]) & (df[feature+"_PC"] == df[feature+"_PC"])]
    df_temp = pd.DataFrame(columns = ["Scolarita", "Eta", feature+"_PG"])
    df_temp["Scolarita"] = df_test["Scolarita"]
    df_temp["Eta"] = df_test["Eta"]
    df_temp[feature+"_PG"] = df_test[feature+"_PG"]

    regr.fit(df_temp.values, df_test[feature+"_PC"])

    # Make predictions using the testing set
    res = regr.predict([[scolarita, eta, raw_score]])
    return res
```

```python
def new_column(df, df_clean, feature):
    col_pc = feature + '_PC'
    col_pg = feature + '_PG'
    col_es = feature + '_ESITO'
    col_new = feature
    tresh = const.const_dict["TRESHOLD"]

    for i in range(dim_df):

        if col_pc in df.columns:

            if (df.loc[i, col_pc] == df.loc[i, col_pc]) and (df.loc[i,col_pg] == df.loc[i,col_pg]):
                if abs(df.loc[i, col_pc] - df.loc[i,col_pg]) < tresh:
                    df_clean.loc[i,col_new] = df.loc[i, col_pc]
                else:
                    df_clean.loc[i,col_new] = df.loc[i, col_pg] #----REG

            elif (df.loc[i, col_pc] != df.loc[i, col_pc]) and (df.loc[i,col_pg] == df.loc[i,col_pg]):
                df_clean.loc[i,col_new] = df.loc[i, col_pg] 
            elif (df.loc[i, col_pc] == df.loc[i, col_pc]) and (df.loc[i,col_pg] != df.loc[i,col_pg]):
                df_clean.loc[i,col_new] = df.loc[i, col_pc]

            elif df.loc[i,col_es] == df.loc[i,col_es]:
                if df.loc[i,col_es] == "NORMA":
                    df_clean.loc[i,col_new] = (const.const_dict[col_new + "_MAX"] + const.const_dict[col_new + "_CUTOFF"])/2
                elif df.loc[i,col_es] == "DEFICIT":
                    df_clean.loc[i,col_new] = const.const_dict[col_new + "_CUTOFF"]/2
                else:
                    df_clean.loc[i,col_new] = const.const_dict[col_new + "_CUTOFF"]

            # else:
                #Null

        else:
            if df.loc[i,col_pg] == df.loc[i,col_pg]:
                df_clean.loc[i,col_new] = df.loc[i, col_pg] #-----REG
                
            elif df.loc[i,col_es] == df.loc[i,col_es]:
                if df.loc[i,col_es] == "NORMA":
                    df_clean.loc[i,col_new] = (const.const_dict[col_new + "_MAX"] + const.const_dict[col_new + "_CUTOFF"])/2
                elif df.loc[i,col_es] == "DEFICIT":
                    df_clean.loc[i,col_new] = const.const_dict[col_new + "_CUTOFF"]/2
                else:
                    df_clean.loc[i,col_new] = const.const_dict[col_new + "_CUTOFF"]
            # else:
                #Null

# plt.hist(df_clean['MMSE'], bins = 20)
```

```python
# "Feature selection"
new_column(df, df_clean, "MMSE")
new_column(df, df_clean, "CLOCKTEST")
new_column(df, df_clean, "COPIAFIGURAREY")
new_column(df, df_clean, "PAROLEREYIMM")
new_column(df, df_clean, "PAROLEREYDIFF")
new_column(df, df_clean, "MEMORIAFIGURAREY")
new_column(df, df_clean, "FAB")
new_column(df, df_clean, "FLUENZAVERBFON")
new_column(df, df_clean, "TESTMATRICIATTENTIVE")
```

```python
# df_clean.to_excel("excel.xlsx")
# Missing Value Imputation
imputer = KNNImputer()

# fit on the dataset
imputer.fit(df_clean.drop(["ID", "Sesso"], axis = 1))

# transform the dataset
imp = imputer.transform(df_clean.drop(["ID", "Sesso"], axis = 1))
imp
```

```python
df_imp = pd.DataFrame(columns = df_clean.drop(["ID", "Sesso"], axis = 1).columns, data = imp)

df_imp.insert(0, "ID", df_clean["ID"].values)
df_imp.insert(len(df_imp.columns), "Patologia", df['Patologia'].values)

df_imp.to_csv("ImputationParziale.csv", index = False)
# df_imp.to_csv("ImputationParzialeTest.csv", index = False)
```
