```python
# Clean MMSE data
MMSE_TRESHOLD = 4
MMSE_NORMA = 27.5
MMSE_DEFICIT = 11

for i in range(dim_df):
    
    if(math.isnan(df.loc[i,'MMSE_PG'])):
        # Raw score absent
        if not df.loc[i,'MMSE_ESITO'] != df.loc[i,'MMSE_ESITO']:
            if df.loc[i,'MMSE_ESITO'] == 'NORMA':
                df_clean.loc[i,"MMSE"] = MMSE_NORMA
            elif df.loc[i,'MMSE_ESITO'] == 'DEFICIT':
                df_clean.loc[i,"MMSE"] = MMSE_DEFICIT
       
    else:
        # Raw score present
        if (not math.isnan(df.loc[i,'MMSE_PC'])) and abs(df.loc[i,'MMSE_PC'] - df.loc[i,'MMSE_PG']) <= MMSE_TRESHOLD:
            df_clean.loc[i,"MMSE"] = df.loc[i,'MMSE_PC']
        else:
            df_clean.loc[i,"MMSE"] = df.loc[i,'MMSE_PG']

df_clean['MMSE_ESITO'] = df['MMSE_ESITO']
```
