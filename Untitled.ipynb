```python
# Data Analysis
import pandas as pd

# Data Visualization
from matplotlib import pyplot as plt
from matplotlib.figure import Figure

# Utilities
import numpy as np
import math
import Constants as const
from sklearn import linear_model
from sklearn.impute import KNNImputer

# print(const.const_dict["MMSE_MAX"])
```

```python
#Load dataset
df = pd.read_excel('TrainingSet.xlsx')
# df = pd.read_excel('TestSet.xlsx')
df.head()
```

```python
# Convert Sex attribute in UpperCase
df['Sesso'] = df['Sesso'].str.upper()

# Remove CAP attribute
df = df.drop(columns =['CAP'])

# Remove unlabeled instances
df = df.loc[df["Patologia"] == df["Patologia"]]

dim_df = len(df['ID'])
```

```python
# Missing Value Imputation
imputer = KNNImputer()

# fit on the dataset
cols = [c for c in df.columns if not ('esito' in c.lower())]
df_temp = df[cols].drop(["ID", "Sesso", "Patologia"], axis = 1)

imputer.fit(df_temp)

# transform the dataset
imp = imputer.transform(df_temp)
imp

df_imp = pd.DataFrame(columns = df_temp.columns, data = imp)

df_imp.insert(0, "ID", df["ID"].values)
df_imp.insert(len(df_imp.columns), "Patologia", df['Patologia'].values)
df_imp
df_imp.to_csv("ImputationCompleta.csv", index = False)
# df_imp.to_csv("ImputationCompletaTest.csv", index = False)
```
